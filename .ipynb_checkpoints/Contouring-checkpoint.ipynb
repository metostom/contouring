```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from shapely import geometry as geo
import numpy as np
```

```python
class ContourData():

    """


    """

    def __init__(self,filepath,xcol,ycol,zcol,tcol,crs):
        
        self.filePath = filepath
        self.xcol = xcol
        self.ycol = ycol
        self.zcol = zcol
        self.tcol = tcol
        self.crs = crs
        self.data = pd.read_excel(filepath)

    def subsetData(self):
        
        times = self.data[self.tcol].unique()
        dfsubs = []
        
        for t in times:
            dft = self.data[self.data[self.tcol] == t]
            dfsubs.append(dft)
        
        self.dfTimes = pd.DataFrame(
                        columns = ['DateTime','DataFrames'],
                        data = np.array([times,dfsubs]).T
                        )
        
    def contourIntervals(self,zvalues,interval=0.25):
        
        min_ = np.floor(np.min(zvalues))
        max_ = np.ceil(np.max(zvalues))
        vals = np.arange(min_,max_,interval)
    
        return vals
        
    def contourData(self,interval):
        
        """
        Create all figure/contour objects using the plt.contour
        
        """
        
        csc = []
        
        for index, row in self.dfTimes.iterrows():
            
            x = row['DataFrames'][self.xcol].values
            y = row['DataFrames'][self.ycol].values
            z = row['DataFrames'][self.zcol].values
            vals = self.contourIntervals(z,interval)
            cs = plt.tricontour(x,y,z,levels=vals)
            csc.append(cs)
        
        self.dfContours = self.dfTimes.copy()
        self.dfContours['CS'] = csc
    
    def extractGeometry(self):
        
        cntrdfs = []
        
        for index, row in self.dfContours.iterrows():
            
            cs = row['CS']
            linestrings = []
            
            for clc in cs.collections:    
                paths = clc.get_paths()[0].vertices
                ls = geo.LineString(paths)
                linestrings.append(ls)
            
            times = [row['DateTime'] for i in range(len(cs.levels))]
            cntrdf = gpd.GeoDataFrame(
                        data = [cs.levels,times,linestrings],
                        )
            cntrdf = cntrdf.T
            cntrdf.columns = ['Contour','DateTime','geometry']
            cntrdfs.append(cntrdf)
        
        self.dfContours['gdf'] = cntrdfs
    
    def mergeGeometry(self):
        
        gdfs = self.dfContours['gdf'].tolist()
        geos = pd.concat(gdfs)
        geo = gpd.GeoDataFrame(geo)
        geo.crs = self.crs
        self.MergedGeo = geos 
        
    def reproject(self,crsNew=4326):
        
        self.MergedGeo.to_crs(crsNew)

    def toGeoJSON(self,filepath):
        
        self.MergedGeo.to_file(filepath)
    

class WebAppCall():
    
    def __init__(self,)
        
    
    
```

```python
cd = ContourData(
        'WellElevationSubset.xlsx',
        'X',
        'Y',
        "Water Elevation(ft NAVD88)",
        "DateTimeRounded",
        "4326"
)
```

```python
cd.subsetData()
cd.contourData(interval=0.25)
cd.extractGeometry()
cd.mergeGeometry()
```
